# DATA INGESTION


```python
import pandas as pd
```

```python
data_to_ingest = pd.read_csv("Data/vt_cd_events.csv")
data_to_ingest
```

```python
print(data_to_ingest.user_coordinates)
```

```python
data_to_ingest = data_to_ingest[data_to_ingest.user_coordinates.isnull() == False]
data_to_ingest
```

```python
data_to_ingest.user_verified.replace({True : "true", False : "false"}, inplace=True)
data_to_ingest.is_retweet.replace({True : "true", False : "false"}, inplace=True)
data_to_ingest
```

### modify the coordinates

```python
def to_string(x):
    x = x.replace("[","")
    x = x.replace("]","")
    a = x.split(",")
    for i in range(len(a)):
        a[i]= float(a[i])
        
    return a


data_to_ingest.user_coordinates = data_to_ingest.user_coordinates.apply(lambda x: to_string(x))
data_to_ingest
```

### handle hashtags

```python
def better_hashtags(x):
    if isinstance(x, str):
        x = x.replace("[","")
        x = x.replace("]","")
        a = x.split(",")
        for i in range(len(a)):
            a[i]= str(a[i])
            a[i] = a[i].strip(" ")
            
            
    else:
        a = ""
    return a

data_to_ingest.hashtags = data_to_ingest.hashtags.apply(lambda x: better_hashtags(x))
data_to_ingest
```

```python
data_to_ingest.drop(columns = ["user_coordinates2"], inplace=True)
```

```python
data_to_ingest.date = data_to_ingest.date.apply(lambda x: x.split(" ")[0])
data_to_ingest.user_created = data_to_ingest.user_created.apply(lambda x: x.split(" ")[0])
data_to_ingest.date
```

```python
data_to_ingest.to_json("Data/to_ingest/twitter_data.json", orient = "records", lines = True)
```

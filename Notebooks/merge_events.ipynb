```python
# dataframe
df      =  pd.read_csv("Data/vt_cases_deaths_complete.csv")  
events  = pd.read_csv("Data/salient_events.csv") 
```

```python
df['date'] = pd.to_datetime(df['date']).dt.date
```

```python
events['date'] = pd.to_datetime(events['date'], dayfirst=True).dt.date
events['date_plus_10'] = pd.to_datetime(events['date_plus_10'], dayfirst=True).dt.date

events
```

```python
for index, row in df.iterrows():
    if df.loc[index,'date'] <= events.loc[0,'date_plus_10'] and df.loc[index,'user_country'] == 'United States':
        df.loc[index,'event'] = events.loc[0,'event']
        df.loc[index,'link'] = events.loc[0,'newspaper_link']
    elif df.loc[index,'date'] > events.loc[0,'date_plus_10'] and df.loc[index,'date'] <= events.loc[1,'date_plus_10'] and df.loc[index,'user_country'] == 'United Kingdom':
        df.loc[index,'event'] = events.loc[1,'event']
        df.loc[index,'link'] = events.loc[1,'newspaper_link'] 
    elif df.loc[index,'date'] > events.loc[1,'date_plus_10'] and df.loc[index,'date'] <= events.loc[2,'date_plus_10'] and df.loc[index,'user_country'] == 'United States':
        df.loc[index,'event'] = events.loc[2,'event']
        df.loc[index,'link'] = events.loc[2,'newspaper_link']
    elif df.loc[index,'date'] > events.loc[2,'date_plus_10'] and df.loc[index,'date'] <= events.loc[3,'date_plus_10'] and (df.loc[index,'user_country'] == 'United States' or df.loc[index,'user_country'] == 'United Kingdom'):
        df.loc[index,'event'] = events.loc[3,'event']
        df.loc[index,'link'] = events.loc[3,'newspaper_link']
    elif df.loc[index,'date'] > events.loc[3,'date_plus_10'] and df.loc[index,'date'] <= events.loc[4,'date_plus_10'] and df.loc[index,'user_country'] == 'United States':
        df.loc[index,'event'] = events.loc[4,'event']
        df.loc[index,'link'] = events.loc[4,'newspaper_link']
    elif df.loc[index,'date'] > events.loc[4,'date_plus_10'] and df.loc[index,'date'] <= events.loc[5,'date_plus_10']:
        df.loc[index,'event'] = events.loc[5,'event']
        df.loc[index,'link'] = events.loc[5,'newspaper_link']    
    elif df.loc[index,'date'] > events.loc[5,'date_plus_10'] and df.loc[index,'date'] <= events.loc[6,'date_plus_10'] and df.loc[index,'user_country'] == 'United Kingdom':
        df.loc[index,'event'] = events.loc[6,'event']
        df.loc[index,'link'] = events.loc[6,'newspaper_link']  
    elif df.loc[index,'date'] > events.loc[6,'date_plus_10'] and df.loc[index,'date'] <= events.loc[7,'date_plus_10'] and df.loc[index,'user_country'] == 'India':
        df.loc[index,'event'] = events.loc[7,'event']
        df.loc[index,'link'] = events.loc[7,'newspaper_link'] 
    elif df.loc[index,'date'] > events.loc[7,'date_plus_10'] and df.loc[index,'date'] <= events.loc[8,'date_plus_10'] and df.loc[index,'user_country'] == 'United Kingdom':
        df.loc[index,'event'] = events.loc[8,'event']
        df.loc[index,'link'] = events.loc[8,'newspaper_link'] 
    elif df.loc[index,'date'] > events.loc[8,'date_plus_10'] and df.loc[index,'date'] <= events.loc[9,'date_plus_10'] and df.loc[index,'user_country'] == 'India':
        df.loc[index,'event'] = events.loc[9,'event']
        df.loc[index,'link'] = events.loc[9,'newspaper_link'] 
    elif df.loc[index,'date'] > events.loc[9,'date_plus_10'] and df.loc[index,'date'] <= events.loc[10,'date_plus_10'] and (df.loc[index,'user_country'] == 'United States' or df.loc[index,'user_country'] == 'United Kingdom'):
        df.loc[index,'link'] = events.loc[10,'newspaper_link']
    else:
        df.loc[index,'event'] = events.loc[10,'event']
        df.loc[index,'link'] = events.loc[10,'newspaper_link']
```

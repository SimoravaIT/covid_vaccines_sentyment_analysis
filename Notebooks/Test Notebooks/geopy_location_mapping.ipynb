```python
import pandas as pd
df = pd.read_csv('Data/vaccination_all_tweets_cleaned.csv', parse_dates=['user_created', 'date'])
```

```python
df
```

```python
unique_types = []
location_list = df['user_location'].tolist()
for x in location_list:
    if x not in unique_types:
        unique_types.append(x)
unique_types
```

```python
def do_geocode(address):
    geopy = Nominatim(user_agent="app")
    try:
        return geopy.geocode(address)
    except GeocoderTimedOut:
        return do_geocode(address)
    
location = do_geocode("La Crescenta-Montrose, CA")
print(location.latitude)
```

```python
location_list = df['user_location'].tolist()
location_list
```

```python
latitude_list = []
for x in location_list: 
    location = do_geocode(x)
    if location is not None:
        print(location.latitude)
        latitude_list.append(location.latitude)
    else:
        latitude_list.append("uknown")
latitude_list
```

```python
df['latitude'] = df['user_location'].apply(lambda x: do_geocode(x))
```

```python
geolocator = Nominatim(user_agent="app")
df['latitude'] = df['user_location'].apply(lambda x: 'uknown' if geolocator.geocode(x) is None else geolocator.geocode(x).latitude)
```

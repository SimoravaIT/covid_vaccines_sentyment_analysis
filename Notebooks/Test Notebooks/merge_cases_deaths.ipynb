```python
def filter_date_range(data, string_name_column_date):
    
    start_date = "2020-12-12"
    end_date = "2021-03-14"

    after_start_date = data[string_name_column_date] >= start_date
    before_end_date = data[string_name_column_date] <= end_date
    between_two_dates = after_start_date & before_end_date
    filtered_dates = data.loc[between_two_dates]
    
    return filtered_dates
```

```python
filt_df = filter_date_range(df, 'date')
filt_cases = filter_date_range(cases, 'Date')
filt_deaths = filter_date_range(deaths, 'Date')
```
